```python
def fib_fast(n):
    dp = [0, 1]
    for i in range(2, n+1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

print(fib_fast(6))
"""Diptiranjan - 2341014099"""
```

```python
import random

def quickselect(arr, k):
    pivot = random.choice(arr)

    left = [x for x in arr if x > pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x < pivot]

    if k <= len(left):
        return quickselect(left, k)
    elif k <= len(left) + len(mid):
        return pivot
    else:
        return quickselect(right, k - len(left) - len(mid))

arr = [3, 2, 1, 5, 6, 4]
print(quickselect(arr, 2))
"""Diptiranjan - 2341014099"""
```

```python
import math

def has_antipodal(points):
    angles = [math.atan2(y, x) for x, y in points]
    angles.sort()

    angle_set = set(angles)

    for angle in angles:
        opposite = (angle + math.pi) % (2 * math.pi)
        if opposite in angle_set:
            return True
    return False
"""Diptiranjan - 2341014099"""
```

```python
def partition(arr, low, high):
    pivot = arr[low]
    i = low + 1
    j = high

    while True:
        while i <= j and arr[i] <= pivot:
            i += 1

        while i <= j and arr[j] >= pivot:
            j -= 1

        if i <= j:
            arr[i], arr[j] = arr[j], arr[i]
        else:
            break

    arr[low], arr[j] = arr[j], arr[low]
    return j


def quick_sort_helper(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)
        quick_sort_helper(arr, low, pivot_index - 1)
        quick_sort_helper(arr, pivot_index + 1, high)


def quick_sort(arr):
    quick_sort_helper(arr, 0, len(arr) - 1)

arr = [37, 2, 6, 4, 89, 8, 10, 12, 68, 45]

print("Original Array:", arr)

quick_sort(arr)

print("Sorted Array:", arr)

print("\nDiptiranjan - 2341014099")
```

```python
def selection_sort(data):
    arr = data[:]
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j][1] < arr[min_idx][1]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr


def bubble_sort(data):
    arr = data[:]
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j][1] > arr[j+1][1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


def insertion_sort(data):
    arr = data[:]
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j][1] > key[1]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr


people = [
    ("Elon Musk", 433.9),
    ("Jeff Bezos", 239.4),
    ("Mark Zuckerberg", 211.8),
    ("Larry Ellison", 204.6),
    ("Bernard Arnault & Family", 181.3),
    ("Larry Page", 161.4)
]

sel_sorted = selection_sort(people)
bub_sorted = bubble_sort(people)
ins_sorted = insertion_sort(people)

def to_dict(sorted_list):
    return {name: worth for name, worth in sorted_list}

print("Selection Sort:", to_dict(sel_sorted))
print("Bubble Sort:", to_dict(bub_sorted))
print("Insertion Sort:", to_dict(ins_sorted))

print("\nDiptiranjan - 2341014099")
```

```python
def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


arr = ['apple', 'orange', 'banana', 'grape']
sorted_arr = merge_sort(arr)

print("Sorted List:", sorted_arr)

print("\nDiptiranjan - 2341014099")
```

```python
def merge_sorted_lists(list1, list2):
    i = j = 0
    result = []

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1

    result.extend(list1[i:])
    result.extend(list2[j:])

    return result

list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]

merged = merge_sorted_lists(list1, list2)

print("Merged List:", merged)

print("\nDiptiranjan - 2341014099")
```
