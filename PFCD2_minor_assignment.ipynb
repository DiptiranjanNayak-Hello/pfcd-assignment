### 1. Significance of Classes
Classes are the foundation of object-oriented programming (OOP) in Python. They:

- Encapsulate data (attributes) and behavior (methods).
- Promote reusability and modularity.
- Enable inheritance and polymorphism, making code extensible.
- Allow modeling of real-world entities (e.g., BankAccount, Book).

### 2. Basic Bank Account Class

```python
class BankAccount:
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient funds!")

    def display_balance(self):
        print(f"Current Balance for {self.owner}: {self.__balance}")

acc = BankAccount("Student_User", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.display_balance()
'''Diptiranjan - 2341014099'''
```

### 3. Book and Chapter Composition

```python
class Chapter:
    def __init__(self, title, page_count):
        self.title = title
        self.page_count = page_count

class Book:
    def __init__(self, title):
        self.title = title
        self.chapters = []

    def add_chapter(self, title, pages):
        self.chapters.append(Chapter(title, pages))

    def get_total_pages(self):
        return sum(chap.page_count for chap in self.chapters)

my_book = Book("Python Basics")
my_book.add_chapter("Variables", 15)
my_book.add_chapter("Loops", 20)
my_book.add_chapter("OOP", 30)

print(f"Book: {my_book.title}")
print(f"Total Page Count: {my_book.get_total_pages()}")
'''Diptiranjan - 2341014099'''
```

### 7. Card Game Simulation

```python
import random
class Card:
    def __init__(self, suit, rank):
        self.suit, self.rank = suit, rank
    def __repr__(self):
        return f"{self.rank} of {self.suit}"

class Deck:
    def __init__(self):
        self.cards = [Card(s, r) for s in ['Hearts', 'Spades'] for r in ['A', 'K', 'Q']]
    def shuffle(self):
        random.shuffle(self.cards)
    def deal(self, num):
        hand = self.cards[:num]
        self.cards = self.cards[num:]
        return hand

class Player:
    def __init__(self, name):
        self.name = name
        self.hand = []

deck = Deck()
deck.shuffle()
players = [Player("Alice"), Player("Bob")]
for p in players:
    p.hand = deck.deal(2)
    print(f"{p.name}'s hand: {p.hand}")

'''Diptiranjan - 2341014099'''
```

### 8. Vehicle and Car Inheritance

```python
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model
    def display_info(self):
        print(f"Vehicle: {self.make} {self.model}")

class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors
    def display_info(self):
        super().display_info()
        print(f"Doors: {self.num_doors}")

v = Vehicle("Toyota", "Camry")
c = Car("Tesla", "Model 3", 4)
v.display_info()
c.display_info()

'''Diptiranjan - 2341014099'''
```

### 9. Shape Polymorphism

```python
class Shape:
    def area(self): pass

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14 * self.r ** 2

class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self): return self.w * self.h

shapes = [Circle(5), Rectangle(4, 5)]
for s in shapes:
    print(f"Area: {s.area()}")

'''Diptiranjan - 2341014099'''
```

### 10. CommissionEmployee with Properties

```python
class CommissionEmployee:
    def __init__(self, first_name, last_name, sales, rate):
        self.first_name = first_name
        self.last_name = last_name
        self.sales = sales
        self.rate = rate

    @property
    def sales(self):
        return self._sales

    @sales.setter
    def sales(self, value):
        if value < 0:
            raise ValueError("Sales cannot be negative!")
        self._sales = value

    def earnings(self):
        return self.sales * self.rate

    def __repr__(self):
        return f"{self.first_name} {self.last_name}: Earnings ${self.earnings()}"

emp = CommissionEmployee("John", "Doe", 5000, 0.1)
print(emp)
emp.sales = 7000
print(f"Updated Earnings: ${emp.earnings()}")

try:
    emp.sales = -100
except ValueError as e:
    print(f"Error: {e}")

'''Diptiranjan - 2341014099'''
```

### 11. Duck Typing

### 12. Operator Overloading

```python
class Complex:
    def __init__(self, r, i):
        self.r, self.i = r, i
    def __add__(self, other):
        return Complex(self.r+other.r, self.i+other.i)
    def __repr__(self):
        return f"{self.r}+{self.i}j"

c1, c2 = Complex(2,3), Complex(4,5)
print(c1 + c2)

'''Diptiranjan - 2341014099'''
```

### 13. Custom Exception for Insufficient Funds

```python
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Withdrawal of {amount} failed. Balance is {balance}.")

class Bank:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        self.balance -= amount
        print("Withdrawal successful.")

my_bank = Bank(500)
try:
    my_bank.withdraw(1000)
except InsufficientFundsError as e:
    print(e)

'''Diptiranjan - 2341014099'''
```

### 15. DataClass vs NamedTuple

Data Classes offer several advantages over Named Tuples by providing a more powerful and flexible way to define data-holding objects.

Key Advantages of Data Classes:

- **Mutability:** Data classes are mutable by default, allowing you to change attribute values after the object is created, whereas named tuples are immutable.
- **Default Values:** You can easily assign default values to attributes in a data class definition.
- **Type Hinting:** They are designed with PEP 526 type hints in mind, making the code more readable and compatible with static type checkers.
- **Method Support:** You can easily define custom methods, properties, and even use inheritance within a data class, which is more cumbersome with named tuples.
- **Automatic Method Generation:** They automatically generate `__init__`, `__repr__`, and `__eq__` methods based on the defined attributes.

```python
from dataclasses import dataclass

@dataclass
class Card:
    suit: str
    rank: str

    # Functionality: Custom method inside the data class
    def describe(self):
        return f"{self.rank} of {self.suit}"

# Advantage: Mutability (Changing a card's rank if needed)
my_card = Card(suit="Hearts", rank="Ace")
print(f"Original: {my_card.describe()}")

my_card.rank = "King"
print(f"Modified: {my_card.describe()}")

'''Diptiranjan - 2341014099'''
```

### 17. Scope Resolution

```python
species = "Global Species"
class Animal:
    species = "Class Species"
    def __init__(self, species):
        self.species = species
    def display_species(self):
        print("Instance species:", self.species)
        print("Class species:", Animal.species)
        print("Global species:", globals()['species'])

a = Animal("Instance Species")
a.display_species()

'''Diptiranjan - 2341014099'''
```

##### Explanation:
- **Instance Scope:** self.species looks at the object's specific attributes (set to "Instance Species" in __init__).

- **Class Scope:** Animal.species looks at the variable defined inside the class block but outside methods.

- **Global Scope:** globals()['species'] explicitly accesses the variable defined at the top level of the script.

Output:

Instance species: Instance Species

Class species: Class Species

Global species: Global Species

### 18. Convert temperature from Celcius to Kelvin

```python
import pandas as pd
import matplotlib.pyplot as plt

c_to_k = lambda c: c + 273.15

celsius_temps = [0, 10, 20, 30, 40, 50, 100]
kelvin_temps = [c_to_k(c) for c in celsius_temps]

df = pd.DataFrame({
    'Celsius (°C)': celsius_temps,
    'Kelvin (K)': kelvin_temps
})

print("--- Temperature Conversion Table ---")
print(df)

df.plot(x='Celsius (°C)', y='Kelvin (K)', kind='line', marker='o', color='b')
plt.title('Celsius to Kelvin Conversion')
plt.xlabel('Celsius (°C)')
plt.ylabel('Kelvin (K)')
plt.grid(True)
plt.show()
print("\n")
'''Diptiranjan - 2341014099'''
```
